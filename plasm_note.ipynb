```python
def find_relative_maxima(values, reference):
    """
    Find relative maxima in a vector with respect to another vector.

    Parameters:
    - values: NumPy array, the vector for which to find maxima.
    - reference: NumPy array, the reference vector for comparison.

    Returns:
    - relative_maxima_indices: List of indices where relative maxima occur.
    """
    relative_maxima_indices = []

    # Ensure both vectors are of the same size
    if len(values) != len(reference):
        raise ValueError("Vectors must be of the same size.")

    # Check for relative maxima
    for i in range(1, len(values) - 1):
        if values[i] > values[i - 1] and values[i] > values[i + 1] and values[i] > reference[i]:
            relative_maxima_indices.append(i)

    return relative_maxima_indices
```

```python
def plane_residuals(params, points):
    # Function to calculate residuals for least squares optimization
    
    # Plane equation: Ax + By + Cz + D = 0
    A, B, C, D = params
    x, y, z = points
    
    # Calculate residuals
    residuals = A*x + B*y + C*z + D
    
    return residuals

def fit_plane(points):
    # Initial guess for plane parameters (you can change this if needed)
    initial_params = [0.1, 0.1, 0.1, 0.1]
    
    # Use least squares optimization to fit the plane
    result = least_squares(plane_residuals, initial_params, args=(points,))
    
    # Extract the optimized parameters
    optimized_params = result.x
    
    return optimized_params
```
